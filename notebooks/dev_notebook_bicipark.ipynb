```python
def to_mercator(lat, long):
    # transform latitude/longitude data in degrees to pseudo-mercator coordinates in metres
    c = gpd.GeoSeries([Point(lat, long)], crs=4326)
    c = c.to_crs(3857)
    return c
```

```python
def distance_meters(lat_start, long_start, lat_finish, long_finish):
    # return the distance in metres between to latitude/longitude pair points in degrees 
    # (e.g.: Start Point -> 40.4400607 / -3.6425358 End Point -> 40.4234825 / -3.6292625)
    start = to_mercator(lat_start, long_start)
    finish = to_mercator(lat_finish, long_finish)
    return start.distance(finish)
```

```python
#Añado a cada data frame limpio una columna de 1 para luego poder hacer un merge
```

```python
merge_unis_bicipark = pd.merge(universidades_clean.assign(key=1), bicipark_stations_clean_rename.assign(key=1), on='key').drop('key', axis=1)
```

```python
merge_unis_bicipark
```

```python
%%time
merge_unis_bicipark['distance'] = merge_unis_bicipark.apply(lambda row: distance_meters(row['location.latitude'],row['location.longitude'],
                                                                                   row['latitude'],row['longitude']), axis = 1)
```

```python
# Por cada universidad, estoy poniendo todas las estaciones de bicimad que tengo. Con eso he calculado las distancias.Con el group by por cada grupo de universidad que me lo ordene de menor a mayor, y te quedas con la primera
```

```python
merge_short_df = merge_unis_bicipark.sort_values(by=['title', 'distance']).reset_index().drop('index', axis=1)
```

```python
minimum_df = merge_short_df.groupby('title').head(1)
```
